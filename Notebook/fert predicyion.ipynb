```python
print(data["Soil Type"].unique())
```

```python
data=data.drop(['Soil Type'],axis=1)
```

```python
data["Crop Type"]=data["Crop Type"].replace("Maize","maize")
data["Crop Type"]=data["Crop Type"].replace("Sugarcane","sugarcane")
data["Crop Type"]=data["Crop Type"].replace("Cotton","cotton")
data["Crop Type"]=data["Crop Type"].replace("Tobacco","tobacco")
data["Crop Type"]=data["Crop Type"].replace("Barley","barley")
data["Crop Type"]=data["Crop Type"].replace("Wheat","wheat")
data["Crop Type"]=data["Crop Type"].replace("Millets","millets")
data["Crop Type"]=data["Crop Type"].replace("Oil seeds","Oil seeds")
data["Crop Type"]=data["Crop Type"].replace("Pulses","pulses")
data["Crop Type"]=data["Crop Type"].replace("Ground Nuts","ground nuts")
```

```python
data
```

```python
import seaborn as sns
plt.figure(figsize=[15,15])
sns.heatmap(data.corr(),annot=True)
```

```python
data=pd.get_dummies(data,columns=["Crop Type"])
```

```python
data
```

```python
x=data.drop(["Fertilizer Name"],axis=1)
y=data["Fertilizer Name"]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
xtrain
```

```python
ytrain
```

```python
ytest
```

```python
xtest
```

```python
ytrain
```

```python
import xgboost
```

```python
XGB=xgboost.XGBClassifier(enable_categorical=True)
XGB.fit(xtrain,ytrain)
pre=XGB.predict(xtest)
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(pre,ytest))
print(classification_report(pre,ytest))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(criterion='entropy',n_estimators=1000)
rfc.fit(xtrain,ytrain)
pre=rfc.predict(xtest)
print(accuracy_score(pre,ytest))
print(classification_report(pre,ytest))
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(xtrain,ytrain)
pre=gnb.predict(xtest)
print(accuracy_score(pre,ytest))
print(classification_report(pre,ytest))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
pre=dt.predict(xtest)
print(accuracy_score(pre,ytest))
print(classification_report(pre,ytest))
```
